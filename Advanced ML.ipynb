```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generating a synthetic dataset
np.random.seed(0)
x1 = np.random.normal(0, 1, 100)
x2 = np.random.normal(0, 1, 100)
y = np.random.choice([0, 1], 100)

# Plotting function
def plot_decision_boundary(w0, w1, w2):
    plt.scatter(x1, x2, c=y, cmap='viridis', edgecolors='k')
    if w2 != 0:
        x_values = np.array(plt.gca().get_xlim())
        y_values = -(w1 / w2) * x_values - w0 / w2
        plt.plot(x_values, y_values, '--')
    else:
        x_values = np.full(100, -w0 / w1)
        plt.scatter(x_values, np.linspace(min(x2), max(x2), 100), color='red')
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.title('Decision Boundary')
    plt.show()

# Parameters (w0, w1, w2)
params = [(0, 1, 1), # Regularizing w0
          (1, 0, 1), # Regularizing w1
          (1, 1, 0)] # Regularizing w2

for param in params:
    plot_decision_boundary(*param)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

# Example synthetic data
np.random.seed(42)
x2 = np.random.normal(0, 1, 100)
y = np.random.choice([0, 1], 100)
w0, w2 = -1, 2  # Example coefficients

# Decision boundary calculation
db_x2 = -w0 / w2

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(x2, y, c=y, cmap='viridis', edgecolors='k', label='Data points')
plt.axvline(x=db_x2, color='r', linestyle='--', label=f'Decision Boundary $x_2 = {db_x2:.2f}$')
plt.xlabel('$x_2$')
plt.ylabel('Probability ($y = 1$)')
plt.title('Decision Boundary Visualization')
plt.legend()
plt.show()

# Classification error calculation
predicted_y = (x2 > db_x2).astype(int)
accuracy = accuracy_score(y, predicted_y)
print(f'Accuracy: {accuracy*100:.2f}%')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

# Example synthetic data
np.random.seed(42)
x2 = np.random.normal(0, 1, 100)
y = np.random.choice([0, 1], 100)

# Parameters to vary
w0_values = [-1, 0, 1]  # Different intercepts
w2_values = [1, 2, 3]   # Different slopes

fig, axes = plt.subplots(nrows=1, ncols=len(w0_values), figsize=(18, 6), sharey=True)

for ax, w0, w2 in zip(axes, w0_values, w2_values):
    # Decision boundary calculation
    db_x2 = -w0 / w2

    # Plotting
    ax.scatter(x2, y, c=y, cmap='viridis', edgecolors='k', alpha=0.7)
    ax.axvline(x=db_x2, color='r', linestyle='--', label=f'Decision Boundary $x_2 = {db_x2:.2f}$')
    ax.set_xlabel('$x_2$')
    ax.set_title(f'$w_0 = {w0}, w_2 = {w2}$')
    ax.legend()

axes[0].set_ylabel('Probability ($y = 1$)')
plt.suptitle('Decision Boundary Visualization Across Different Parameters')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Example of calculating accuracy for one scenario (you can loop this part as well)
predicted_y = [(x2 > -w0 / w2).astype(int) for w0, w2 in zip(w0_values, w2_values)]
accuracies = [accuracy_score(y, pred) for pred in predicted_y]
for w0, w2, acc in zip(w0_values, w2_values, accuracies):
    print(f'Accuracy for w0={w0}, w2={w2}: {acc*100:.2f}%')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

# Example synthetic data
np.random.seed(42)
x1 = np.random.normal(0, 1, 100)  # Feature x1
x2 = np.random.normal(1, 2, 100)  # Feature x2
y = np.random.choice([0, 1], 100)  # Binary target variable

# Parameters to vary
w0_values = [-1,0.5 , 0, 1]  # Different intercepts
w2_values = [0.5, 1, 2, 3]   # Different slopes for x2

fig, axes = plt.subplots(nrows=1, ncols=len(w0_values), figsize=(18, 6), sharey=True)

for ax, w0, w2 in zip(axes, w0_values, w2_values):
    # Decision boundary calculation
    db_x2 = -w0 / w2

    # Plotting
    scatter = ax.scatter(x1, x2, c=y, cmap='viridis', edgecolors='k', alpha=0.7)
    ax.axhline(y=db_x2, color='r', linestyle='--', label=f'Decision Boundary $x_2 = {db_x2:.2f}$')
    ax.set_xlabel('$x_1$')
    ax.set_title(f'$w_0 = {w0}, w_2 = {w2}$')
    ax.legend()

axes[0].set_ylabel('$x_2$')
plt.suptitle('Decision Boundary Visualization Across Different Parameters')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

# Creating colorbar
fig.colorbar(scatter, ax=axes.ravel().tolist(), label='Probability ($y = 1$)')

plt.show()

# Calculate accuracy for one scenario (this can be looped as well)
predicted_y = [(x2 > db_x2).astype(int) for db_x2 in (-np.array(w0_values) / np.array(w2_values))]
accuracies = [accuracy_score(y, pred) for pred in predicted_y]
for w0, w2, acc in zip(w0_values, w2_values, accuracies):
    print(f'Accuracy for w0={w0}, w2={w2}: {acc*100:.2f}%')
```
